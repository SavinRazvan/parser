```python
import subprocess

# Specify the file number to be processed
file_number = 10

# Construct the file path for the sentence file
file_path = f'sentences/{file_number}.txt'

# Run parser.py with the specified sentence file
# The subprocess.run function executes the parser.py script with the given file as an argument.
# capture_output=True captures the standard output and standard error.
# text=True ensures that the output is returned as a string.
result = subprocess.run(['python', 'parser.py', file_path], capture_output=True, text=True)

# Check if the subprocess completed successfully
if result.returncode == 0:
    # Print the standard output from the parser script
    print(result.stdout)
else:
    # Print an error message if the script did not run successfully
    print(f"Error: {result.stderr}")
```
